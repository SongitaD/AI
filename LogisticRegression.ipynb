```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Example data: Hours studied VS Exam Outcome (0=fail; 1=pass)
# Data: [Hours Studied]
X=np.array([[1],[2],[3],[4],[5],[6],[7]]);

#Labels: Exam Outcome
y=np.array([0,0,0,1,1,1,1]);

# Create and train the logistic regression model
model = LogisticRegression()
model.fit(X, y)

# Predict the probability and class for a new data point (e.g., 4.5 hours)
hours_new = np.array([[4.5]])
probability = model.predict_proba(hours_new)[0][1]  # probability of passing (class 1)
prediction = model.predict(hours_new)[0]


print("Probability of passing for 4.5 hours of study:", probability)
print("Predicted outcome (0 = fail, 1 = pass):", prediction)

# Plot the data and logistic regression curve
x_values = np.linspace(0, 8, 100).reshape(-1, 1)
y_prob = model.predict_proba(x_values)[:, 1]

plt.scatter(X, y, color='blue', label="Data points")
plt.plot(x_values, y_prob, color='red', label="Logistic regression curve")
plt.xlabel("Hours Studied")
plt.ylabel("Probability of Passing")
plt.title("Logistic Regression: Exam Passing Prediction")
plt.legend()
plt.show()
```

**Explanation:**

**Data Preparation:**
We define X as the number of hours studied and y as the exam outcome (0 for fail, 1 for pass).

**Model Training:**
We create a logistic regression model using scikit-learn and train it on our data with model.fit(X, y).

**Prediction: **
The model predicts the probability of passing for a student who studies 4.5 hours, using model.predict_proba(), and also gives a binary prediction with model.predict().

**Visualization:**
We plot the original data points along with the logistic regression curve, which shows the probability of passing as a function of study hours.
