```python
#Linear Regression Examples
#Below is a plot illustrating our simple linear regression example:
#Blue dots: Represent the data points for students:
#(1, 50), (2, 60), (3, 70)
#Red line: Represents the best-fit line using the equation
#Score = 40 + 10 Ã— (Hours Studied)
#This line helps us predict a student's score based on the number of hours they studied.

import matplotlib.pyplot as plt
import numpy as np

# Data points
x = np.array([1, 2, 3])
y = np.array([50,60,70])

# Regression line: y = 40 + 10*x
x_line = np.linspace(0,5,100)
y_line=40 +10*x_line

# Plotting
plt.figure(figsize=(6,4))
plt.scatter(x,y,color='blue',label='Data Points')
plt.plot(x_line,y_line,color='red',label='Regression Line')
plt.xlabel('Hours Studied (X)')
plt.ylabel('Score (Y)')
plt.title('Linear Regression: Predicting Score from Hours Studied')
plt.legend()
plt.grid(True)
plt.show()


```
